```python
from transformers import pipeline
```

```python
a = pipeline("sentiment-analysis")
```

```python
text = input("Enter the text:")
t = a(text,max_length=200)
```

```python
gen = pipeline("text-generation")
text1 = input("Enter the text: ")
t = gen(text1 , max_length = 200)
print(t[0].get('generated_text'))
```

```python
qa = pipeline("question-answering")
```

```python
c = '''Bitty bought  some butter, but the butter was bitter, so Bity bought some better butter.
so, Bity should buy better butter.to make bitteer butter better'''
```

```python
q = "What does Bitty bought"
```

```python
result = qa({'question':q,'context':c})
print(result)
```
