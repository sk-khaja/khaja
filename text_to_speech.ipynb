```python
from gtts import gTTS
from IPython.display import Audio
```

```python
text = "Hello! guys welcome to google colab"
```

```python
language = 'en'
```

```python
tts = gTTS(text=text, lang=language)
```

```python
tts.save("output.mp3")
```
